### Ejercicio distribución Normal

Supongamos que las puntuaciones de un examen de inglés siguen una distribución normal con una media (μ) de 85 y una desviación estándar (σ) de 7.

¿Cuál es la probabilidad de que un estudiante obtenga una puntuación menor a 90?


```python
from scipy.stats import norm

# Datos dados
mu = 85  # media
sigma = 7  # desviación estándar

# Calcular z para x = 90
z_90 = (90 - mu) / sigma
prob_menor_90 = norm.cdf(z_90) * 100  # multiplicado por 100 para obtener el porcentaje

# Imprimir los resultados
print(f"La probabilidad de obtener una puntuación menor a 90 es aproximadamente: {prob_menor_90:.2f}%")

```

### Ejercicio

Supongamos que las alturas de una población siguen una distribución normal con una media (μ) de 170 cm y una desviación estándar (σ) de 5 cm.

Encuentra la probabilidad de que una persona seleccionada al azar tenga una altura menor a 175 cm

¿Cuál es la probabilidad de que una persona tenga una altura mayor a 180 cm?

¿Cuál es la probabilidad de que una persona tenga una altura entre 165 cm y 175 cm?

```python
from scipy.stats import norm

# Definir los parámetros dados
mu = 170  # media en cm
sigma = 5  # desviación estándar en cm

# Función para calcular la probabilidad utilizando la distribución normal estándar
def probabilidad_distribucion_normal(x, mu, sigma):
    z = (x - mu) / sigma
    return norm.cdf(z)

# 1. Probabilidad de que una persona tenga una altura menor a 175 cm
prob_menor_175 = probabilidad_distribucion_normal(175, mu, sigma) * 100  # multiplicado por 100 para obtener el porcentaje

# 2. Probabilidad de que una persona tenga una altura mayor a 180 cm
prob_mayor_180 = (1 - probabilidad_distribucion_normal(180, mu, sigma)) * 100  # multiplicado por 100 para obtener el porcentaje

# 3. Probabilidad de que una persona tenga una altura entre 165 cm y 175 cm
prob_entre_165_y_175 = (probabilidad_distribucion_normal(175, mu, sigma) - probabilidad_distribucion_normal(165, mu, sigma)) * 100  # multiplicado por 100 para obtener el porcentaje

# Imprimir resultados
print(f"1. La probabilidad de que una persona tenga una altura menor a 175 cm es aproximadamente: {prob_menor_175:.2f}%")
print(f"2. La probabilidad de que una persona tenga una altura mayor a 180 cm es aproximadamente: {prob_mayor_180:.2f}%")
print(f"3. La probabilidad de que una persona tenga una altura entre 165 cm y 175 cm es aproximadamente: {prob_entre_165_y_175:.2f}%")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Definir los parámetros dados
mu = 170  # media en cm
sigma = 5  # desviación estándar en cm

# Crear un rango de valores para las alturas
x = np.linspace(150, 190, 1000)

# Calcular la función de densidad de probabilidad (FDP)
pdf = norm.pdf(x, mu, sigma)

# Graficar la FDP
plt.plot(x, pdf, label='FDP')

# Rellenar el área para la probabilidad de altura menor a 175 cm
x_menor_175 = np.linspace(150, 175, 1000)
plt.fill_between(x_menor_175, norm.pdf(x_menor_175, mu, sigma), color='blue', alpha=0.5, label='P(X < 175)')

# Añadir etiquetas y leyenda
plt.title('Distribución Normal')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

# Mostrar el gráfico
plt.show()
```

### Ejercicio distribución chi-cuadrado

##### Ejercicio 1:

Si consideramos una distribución chi-square con cuatro grados de libertad, cual es el valor que deja solo un 1% de probabilidad en la cola derecha en la gráfica de densidad de probabilidad.

```python
from scipy.stats import chi2

grados_libertad = 4
probabilidad = 0.01

# Calcular el valor crítico
valor_critico = chi2.ppf(1 - probabilidad, df=grados_libertad)

valor_critico
```

##### Ejercicio 1:

Queremos determinar el valor crítico de la distribución t de Student con 7 grados de libertad y una probabilidad del 5% de encontrar un valor igual o mayor.

```python
from scipy.stats import t

grados_libertad = 7
probabilidad = 0.05

# Calculamos el valor crítico
valor_critico = t.ppf(1 - probabilidad, df=grados_libertad)

valor_critico
```

##### Función de densidad de probabilidad

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

grados_libertad = 7

# Generar datos para el eje x
x = np.linspace(-5, 5, 1000)
# Calcular la función de densidad de probabilidad
pdf = t.pdf(x, df=grados_libertad)

# Graficar la densidad de probabilidad
plt.plot(x, pdf, label=f'Distribución t con {grados_libertad} grados de libertad')
plt.title('Densidad de Probabilidad de la Distribución t')
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```
