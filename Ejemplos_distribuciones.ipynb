### Ejemplo distribuciones de poisson

Supongamos que en un call center se recibe un promedio de 12 llamadas telefónicas por hora. Queremos calcular la probabilidad de que, durante un período de 30 minutos, se reciban exactamente 6 llamadas telefónicas.

Pregunta: ¿Cuál es la probabilidad de recibir exactamente 6 llamadas en un período de 30 minutos?

```python
from scipy.stats import poisson

# Parámetros de la distribución de Poisson
lmbda = 12  # Promedio de llamadas por hora
t = 0.5    # 30 minutos en horas

# Calcula la probabilidad usando la función de masa de probabilidad (PMF) de Poisson
probabilidad = poisson.pmf(6, lmbda * t)

print(f"La probabilidad de recibir exactamente 6 llamadas en 30 minutos es: {probabilidad:.4f}")
```

#### Función de masa de probabilidad

```python
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parámetros de la distribución de Poisson
lmbda = 12  # Tasa de llamadas en 1 hora
t = 0.5 # 30 minutos en horas

# Valores de k desde 0 hasta 15 (para visualizar más allá de 6 llamadas)
k_values = range(0, 16)

# Calcula la PMF para los valores de k dados
pmf_values = poisson.pmf(k_values, lmbda*t)

# Crea el gráfico de la PMF
plt.bar(k_values, pmf_values, align='center', alpha=0.7)
plt.plot(k_values, pmf_values, '--')
plt.xlabel('Número de Llamadas')
plt.ylabel('Probabilidad')
plt.title('Función de Masa de Probabilidad (PMF) - Distribución de Poisson')
plt.xticks(k_values)
plt.show()
```

### Ejercicio de Distribución Binomial

Un fabricante de bombillas eléctricas afirma que el 90% de sus bombillas son duraderas y funcionan correctamente. Un inspector selecciona aleatoriamente 10 bombillas de un lote y las prueba.

Pregunta: ¿Cuál es la probabilidad de que exactamente 8 de las 10 bombillas seleccionadas funcionen correctamente?

```python
from scipy.stats import binom

# Parámetros de la distribución
n = 10  # Tamaño de la muestra (10 bombillas)
p = 0.90  # Probabilidad de éxito (90% de probabilidad de que una bombilla funcione correctamente)
k = 8  # Número de éxitos deseados (8 bombillas funcionando correctamente)

# Calcular la probabilidad usando la función de masa de probabilidad (PMF) de la distribución binomial
probabilidad = binom.pmf(k, n, p)

# Imprimir el resultado
print(f"La probabilidad de que exactamente {k} de las {n} bombillas funcionen correctamente es: {probabilidad:.4f}")
```

### Ejemplo de distribución uniforme

Supongamos que en una feria, un juego de azar ofrece un premio en efectivo. El juego consiste en lanzar un dado equilibrado de seis caras. Si el número obtenido es par (2, 4 o 6), el jugador gana $20. Si el número es impar (1, 3 o 5), el jugador pierde $10.

Calcular la ganancia esperada para un jugador en este juego.

```python
# Definir las ganancias y probabilidades
ganancia_par = 20
ganancia_impar = -10
probabilidad_par = 1/2
probabilidad_impar = 1/2

# Calcular la ganancia esperada
ganancia_esperada = (ganancia_par * probabilidad_par) + (ganancia_impar * probabilidad_impar)

# Imprimir el resultado
print("La ganancia esperada en el juego es:", ganancia_esperada)
```

```python
import matplotlib.pyplot as plt

# Número de valores posibles
n = 6

# Probabilidad para cada valor (uniforme discreta)
p = 1 / n

# Lista de valores posibles
valores = list(range(1, n+1))

# Lista de probabilidades para cada valor
probabilidades = [p] * n

# Crear el gráfico de barras
plt.bar(valores, probabilidades, align='center', alpha=0.7)
plt.plot(valores, probabilidades, '--')

# Etiquetas para los ejes
plt.xlabel('Valores')
plt.ylabel('Probabilidades')

# Título del gráfico
plt.title('PMF - Distribución Uniforme Discreta')

# Mostrar el gráfico
plt.show()
```
